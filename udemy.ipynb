```python
from bs4 import BeautifulSoup
import base64
import requests
import csv


def get_instructors_data(url):
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text, 'html.parser')

    page_info = soup.find('div', class_ = 'ud-container instructor-profile--page-container--1oU4J')
    instructor_info = soup.find('div', class_ = 'instructor-profile--instructor-info--1-iYp instructor-profile--left-column--1_cdC')
    
    instructor_name = instructor_info.find('h1').text.strip()
    bio = instructor_info.find('div', class_ = 'show-more-module--content--cjTh0').find('p').text.strip()
    stats = instructor_info.find('div', class_ = 'instructor-profile--instructor-stats--1MA1N').find_all('div', class_ ="ud-heading-xl")
    students = stats[0].text.strip()
    reviews = stats[1].text.strip()

    
    #############
    #This is for instructor's image
    data_url = soup.find('img', class_ = 'instructor-profile--instructor-image--1wMLG instructor-profile--right-column--lU2rQ')['src']
    # Remove the "data:image/svg+xml," prefix from the data URL
    encoded_content = data_url.replace('data:image/svg+xml,', '')

    # Calculate the necessary padding length
    padding_length = 4 - (len(encoded_content) % 4)

    # Add padding if needed
    encoded_content += '=' * padding_length

    # Decode the base64-encoded content
    decoded_content = base64.b64decode(encoded_content)
    image = decoded_content
    #################
    
    links = soup.find('div', class_ = 'instructor-profile--social-links--3x2ms').find_all('a')
    sites = {}
    
    for link in links:
        key = link['data-purpose'].split('-')[0]
        value = link['href']
        
        sites[key] = value
    
    websites = list(sites.values())

    data = [instructor_name, bio, reviews, students, image, websites]
    
    return data
    
   
    

urls = ['https://www.udemy.com/user/janis-smilga-3/',
        'https://www.udemy.com/user/4b4368a3-b5c8-4529-aa65-2056ec31f37e/',
        'https://www.udemy.com/user/jonasschmedtmann/',
        'https://www.udemy.com/user/andrei-neagoie/',
        'https://www.udemy.com/user/academind/',
        'https://www.udemy.com/user/moshfeghhamedani/',
        'https://www.udemy.com/user/timbuchalka/',
        'https://www.udemy.com/user/joseportilla/',
        'https://www.udemy.com/user/coltsteele/',
        'https://www.udemy.com/user/sgslo/']

csv_data = [['Name', 'Bio', 'Reviews', 'Students', 'Image', 'Links']]

for url in urls:
    data = get_instructors_data(url)
    csv_data.append(data)

    
 # Define the filename
filename = 'instructors.csv'

# Open the file in 'write' mode
with open(filename, 'w', newline='') as csvfile:
    # Create a CSV writer object
    csvwriter = csv.writer(csvfile)

    # Write the data to the CSV file
    csvwriter.writerows(csv_data)

print(f'CSV file "{filename}" created successfully.')
```
